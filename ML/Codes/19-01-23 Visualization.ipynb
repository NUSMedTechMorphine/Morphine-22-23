### Looking at "Jumps"

```python
fall2_data = fall2_data.to_numpy()
fall3_data = fall3_data.to_numpy()
normal4_data = normal4_data.to_numpy()
normal5_data = normal5_data.to_numpy()
```

```python
fall2_jumps = fall2_data[1:,:] - fall2_data[:-1,:]
fall3_jumps = fall3_data[1:,:] - fall3_data[:-1,:]
normal4_jumps = normal4_data[1:,:] - normal4_data[:-1,:]
normal5_jumps = normal5_data[1:,:] - normal5_data[:-1,:]
```

```python
plt.hist(fall2_jumps[:,0], alpha = 0.5, color = 'blue')
plt.hist(fall2_jumps[:,1], alpha = 0.5, color = 'red')
plt.hist(fall2_jumps[:,2], alpha = 0.5, color = 'green')
plt.show()
```
