```python
os.chdir(home_dir)
all_files = os.listdir()
adl_files = []
fall_files = []
for file in all_files:
    if "adl" in file.lower():
        adl_files.append(file)
    else:
        fall_files.append(file)
```

```python
def convert(s):
    split = str(s).replace(',','.').split('.')
    return_string = ""
    for i, element in enumerate(split):
        if i == 0:
            return_string = element
        elif i == 1:
            return_string = return_string + "." + element
        else:
            return_string = return_string + element
    
    return float(return_string)
```

```python
for file in all_files:
    data = pd.read_csv(file, skiprows=40, sep = ';').iloc[:,:-1]
    M, N = data.shape
    for i in range(M):
        if i == 0: continue
        for j in (2,3,4):
            data.iloc[i, j] = convert(data.iloc[i,j])
    data.to_csv("C:/Users/ernest.liu/Documents/ZZZ Misc/UMAFall_Dataset/UMAFall_Dataset Formatted/" + file[:-4] + "_reformatted.csv")
    print(f"{file} completed")
```
